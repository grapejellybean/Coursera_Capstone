```python
import urllib.request
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urllib.request.urlopen(url)
from bs4 import BeautifulSoup
soup = BeautifulSoup(page, "lxml")
```

```python
soup.title
soup.title.string
```

```python
all_tables=soup.find_all("table")
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
A=[]
B=[]
C=[]
for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['PostalCode'])
df['Borough']=B
df['Neighbourhood']=C
df.head()
```

```python
df = df[df['Borough'] != 'Not assigned\n'].reset_index(drop=True)
df_sort = df.sort_values(by=['PostalCode','Borough'])

df_sort.reset_index(inplace=True)
df_sort.drop('index',axis=1,inplace=True)
df_sort.head()
```

```python
#analyze the neighbourhooods
toronto_onehot=pd.get_dummies(toronto_venues[["Venue Category"]], prefix="", prefix_sep="")
```

```python
toronto_onehot["Neighbourhood"]=toronto_venues["Neighbourhood"]
```

```python
fixed_columns = [toronto_onehot.columns[-1]] + list(toronto_onehot.columns[:-1])
toronto_onehot = toronto_onehot[fixed_columns]

toronto_onehot.head()
```

```python
toronto_onehot.shape
```

```python
#group by neighborhood
toronto_grouped = toronto_onehot.groupby('Neighbourhood').mean().reset_index()
toronto_grouped.head()
```

```python
toronto_grouped.shape
```

```python
#group by frenquency
num_top_venues = 10

for hood in toronto_grouped['Neighbourhood']:
    print("----"+hood+"----")
    temp = toronto_grouped[toronto_grouped['Neighbourhood'] == hood].T.reset_index()
    temp.columns = ['venue','freq']
    temp = temp.iloc[1:]
    temp['freq'] = temp['freq'].astype(float)
    temp = temp.round({'freq': 2})
    print(temp.sort_values('freq', ascending=False).reset_index(drop=True).head(num_top_venues))
    print('\n')
```

```python
dict1={}

for  hood in toronto_grouped["Neighbourhood"]:
    val=[]
    #print(f"-------{hood}----")
    temp=toronto_grouped[toronto_grouped["Neighbourhood"]==hood].T.reset_index()
    temp.columns=["venue","freq"]
    temp=temp[1:]
    temp["freq"]=round(temp["freq"].astype(float),2)
    val=list(temp.sort_values(by="freq",axis=0,ascending=False).reset_index(drop=True).head(4)["venue"])
    dict1[hood]=val
```
