```python
import urllib.request
url = "https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M"
page = urllib.request.urlopen(url)
from bs4 import BeautifulSoup
soup = BeautifulSoup(page, "lxml")
```

```python
soup.title
soup.title.string
```

```python
all_tables=soup.find_all("table")
all_tables
```

```python
right_table=soup.find('table', class_='wikitable sortable')
right_table
```

```python
A=[]
B=[]
C=[]
for row in right_table.findAll('tr'):
    cells=row.findAll('td')
    if len(cells)==3:
        A.append(cells[0].find(text=True))
        B.append(cells[1].find(text=True))
        C.append(cells[2].find(text=True))
```

```python
import pandas as pd
df=pd.DataFrame(A,columns=['PostalCode'])
df['Borough']=B
df['Neighbourhood']=C
df.head()
```

```python
df = df[df['Borough'] != 'Not assigned\n'].reset_index(drop=True)
df_sort = df.sort_values(by=['PostalCode','Borough'])

df_sort.reset_index(inplace=True)
df_sort.drop('index',axis=1,inplace=True)
df_sort.head()
```

```python
df_sort.shape[0]
```

```python
csv_path = 'https://cocl.us/Geospatial_data'
df_lat = pd.read_csv(csv_path)
df_lat.head()
```

```python
df_lat.rename(columns={"Postal Code": "PostalCode"}, inplace=True)
df_lat.head()
```

```python
df_sort['PostalCode'] = df_sort['PostalCode'].str.replace('\n', '')
df_sort['PostalCode'] = df_sort['PostalCode'].str.replace('\n', '')
df_sort['Borough'] = df_sort['Borough'].str.replace('\n', '')
df_sort.head()
```

```python
df_lat.columns=['PostalCode', 'Latitude', 'Longitude']
toronto_df=pd.merge(df_sort,df_lat,how='inner',on="PostalCode")
toronto_df.head()
```

```python
print('The dataframe has {} boroughs and {} neighborhoods.'.format(
        len(toronto_df['Borough'].unique()),
        toronto_df.shape[0]
    )
)
```

```python
address = 'Toronto, ON'

geolocator = Nominatim(user_agent="ON")
location = geolocator.geocode(address)
latitude = location.latitude
longitude = location.longitude
print('The geograpical coordinate of Toronto are {}, {}.'.format(latitude, longitude))
```

```python
df_borough=toronto_df[toronto_df["Borough"].str.contains("Toronto")].reset_index(drop=True)
df_borough["Borough"].unique()
```

```python
df_borough["color"]=df_borough["Borough"].map({'East Toronto':"red", 'Central Toronto':"black", 'Downtown Toronto':"green",
       'West Toronto':"blue"})
```

```python
# create map of Torronto using latitude and longitude values
map_toronto = folium.Map(location=[latitude, longitude], zoom_start=10)
# add markers to map
for lat, lng, borough, neighborhood,clr in zip(df_borough['Latitude'], df_borough['Longitude'], df_borough['Borough'], df_borough['Neighbourhood'],df_borough["color"]):
    label = '{}, {}'.format(neighborhood, borough)
    label = folium.Popup(label, parse_html=True)
    folium.CircleMarker(
        [lat, lng],
        radius=5,
        popup=label,
        color=clr,
        fill=True,
        fill_color='#3186cc',
        fill_opacity=0.7,
        parse_html=False).add_to(map_toronto) 
map_toronto
```
